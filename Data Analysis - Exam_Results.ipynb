# Questions

1. Does taking a test preparation course improve average scores?

```python
df['test preparation course'].unique()
```

```python
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

- Yes students who have completed test preparation course have higher average scores.

2. Which gender is better on average at math?

```python
df.groupby('gender')['math score'].mean()
```

- Male students did better at math on average.

3. Which gender is better on average at all three subjects? 

```python
df['Total Score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
df.groupby('gender')['Total Score'].mean()
```

- Female students are better on average at all subjects.
